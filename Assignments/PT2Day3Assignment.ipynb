```python
import pandas as pd

df = pd.read_csv("C:\\Users\\jdangelo\\SkillStormPandas\\20240617-DE-TS-LectureMaterials\\Python\\Data\\imdb_top_1000.csv")
#Remove useless Columns, I do not use the following
df.drop(columns=['Poster_Link', 'Overview', 'Meta_score', 'No_of_Votes', 'Certificate'], inplace=True)

#Clean Gross (str to int), replacing NaN with 0
df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')
df['Gross'] = df['Gross'].fillna(0).astype(int)
```

```python
#Find oldest, removing any without Years
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df = df.dropna(subset=['Released_Year'])

df.loc[df['Released_Year'].idxmin()]
```

```python
#Find newist
df.loc[df['Released_Year'].idxmax()]
```

```python
#Find top 10
df2 = df.sort_values(by='IMDB_Rating', ascending=False)
df2.head(10)
```

```python
#Top movie in genre based on IMDB
df3 = df.copy()
df3['Genre'] = df3['Genre'].str.split(', ')
df3 = df3.explode('Genre')
#Nested for loop, Outer : genre, Inner Movies
for genre, movies in df3.groupby('Genre'):
   print(f"{genre} : {movies.sort_values(by='IMDB_Rating',ascending=False).iloc[0]['Series_Title']}")


     
```

```python
#Director with most movies
df["Director"].value_counts()
```

```python
#Stars with most movies
stars_temp = df.copy()
stars_temp['Stars'] = stars_temp[['Star1','Star2','Star3','Star4']].values.tolist()
stars_temp.drop(['Star1', 'Star2', 'Star3', 'Star4'], inplace=True,axis=1)
stars_temp = stars_temp.explode('Stars')
stars_temp['Stars'].value_counts().idxmax()
```

```python
#Highest Gross for each Genre
df4 = df.copy()
df4['Genre'] = df4['Genre'].str.split(', ')
df4 = df4.explode('Genre')

for genre, movies in df4.groupby('Genre'):
    print(f"{genre} : {movies.sort_values(by='Gross',ascending=False).iloc[0]['Series_Title']}")

     
```

```python
#Lowest Grossing for each Director
df5 = df.copy()

for director, movies in df5.groupby('Director'):
    print(f"{director} : {movies.sort_values(by='Gross',ascending=False).iloc[0]['Series_Title']}")

     
```
