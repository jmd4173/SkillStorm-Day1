```python
import pandas as pd


df = pd.read_csv("C:\\Users\\jdangelo\\SkillStormPandas\\20240617-DE-TS-LectureMaterials\\Python\\Data\\imdb_top_1000.csv")
df.drop(['Director', 'Poster_Link', 'Certificate', 'Runtime', 'Released_Year', 
         'Star1', 'Star2', 'Star3', 'Star4', 'Overview', 'Meta_score', 
         'Gross', 'No_of_Votes'], axis='columns', inplace=True)
df['Genre'] = df['Genre'].apply(lambda x: x.split(','))
genres = df['Genre'].explode()
one_hot_encoded_genres = pd.get_dummies(genres).groupby(level=0).sum()
one_hot_encoded_genres = one_hot_encoded_genres.groupby(one_hot_encoded_genres.columns, axis=1).sum()
df = df.join(one_hot_encoded_genres)
df.drop('Genre', axis=1, inplace=True)
df['IMDB_Rating'] = pd.to_numeric(df['IMDB_Rating'], errors='coerce')
result = df[df['Action'] == 1].sort_values(by='IMDB_Rating', ascending=False)[['IMDB_Rating', 'Series_Title']].head(1)

print(result)
```
